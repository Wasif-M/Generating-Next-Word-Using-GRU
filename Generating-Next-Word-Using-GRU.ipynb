```python
import pandas as pd
import numpy as np
import tensorflow as tf
import keras
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.models import load_model
```

```python
with open('Text.txt','r',encoding='latin-1') as f:
  data=f.readlines()
```

```python
corpus=data[:5000]
```

```python
text_data=[sentence.replace('\n',' ') for sentence in corpus]
```

```python
text_data = ' '.join(corpus)
```

```python
text_data[200:2000]
```

```python
tokenizer=Tokenizer()
tokenizer.fit_on_texts([text_data])
```

```python
tokenizer.word_index
```

```python
len(tokenizer.word_index)
```

```python
sent='He is very happy man.'
tokenized_sent=tokenizer.texts_to_sequences([sent])
tokenized_sent
```

```python
input_seq=[]
for sent in text_data.split('\n'):
  tokenized_sent=tokenizer.texts_to_sequences([sent])[0]
  for i in range(1,len(tokenized_sent)):
    input_seq.append(tokenized_sent[:i+1])

for i in input_seq:
  print(i)
```

```python
len(input_seq)
```

```python
max_len=max([len(sent) for sent in input_seq])
```

```python
max_len
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
padded_input_seq=pad_sequences(input_seq,maxlen=max_len,padding='pre',truncating='pre')
```

```python
padded_input_seq[0]
```

```python
X = padded_input_seq[:, :-1]
y = padded_input_seq[:, -1]
```

```python
X[:5], X.shape
```

```python
list(y)
```

```python
y, y.shape
```

```python
word2idx=tokenizer.word_index
len_vocab=len(word2idx)
len_vocab
```

```python
y = y.reshape(-1,1)
```

```python
y.shape
```

```python
from tensorflow.keras.utils import to_categorical
y=to_categorical(list(y),num_classes=len_vocab+1)
```

# With Trainable Embedding

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, GRU, Dense

# Creating the Model
model1 = Sequential()
model1.add(Embedding(len_vocab + 1, 30, input_length=max_len - 1))
model1.add(GRU(150, return_sequences=True))  # First GRU layer with return_sequences=True
model1.add(GRU(100))  # Second GRU layer without return_sequences (output the final hidden state)
model1.add(Dense(len_vocab + 1, activation='softmax'))

# Compiling the model
model1.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

# Checking summary
model1.summary()
```

```python
model1.fit(X,y,epochs=50)
```

```python
# Predictor

import time
text = "Arya"

for i in range(30):
  # tokenize
  token_text = tokenizer.texts_to_sequences([text])[0]
  # padding
  padded_token_text = pad_sequences([token_text], maxlen=max_len-1, padding='pre')
  # predict
  pos = np.argmax(model1.predict(padded_token_text))

  for word,index in tokenizer.word_index.items():
    if index == pos:
      text = text + " " + word
      print(text)
      # time.sleep(1)
```

```python
model1.save('model1.keras')
```
